```python
import numpy as np
import pandas as pd

# Load the DataFrame from the CSV file
df = pd.read_csv('/content/1774404110.csv')

unique_cylinders = df['cylinders'].unique()
mean_mpg_by_cylinders = {}

for c in unique_cylinders:
    # Filter mpg for current cylinder count and convert to numpy array
    cyl_mpg = df[df['cylinders'] == c]['mpg'].to_numpy()
    mean_mpg_by_cylinders[c] = np.mean(cyl_mpg)

print("Mean MPG by number of cylinders:")
for cyl, mean_mpg_val in mean_mpg_by_cylinders.items():
    print(f"Cylinders: {cyl}, Mean MPG: {mean_mpg_val:.2f}")
```
